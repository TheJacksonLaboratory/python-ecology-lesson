## Combing DataFrames with pandas

```python
    # Copy these lines to your new notebook:

    %matplotlib inline

    import pandas as pd

    surveys_df = pd.read_csv("https://ndownloader.figshare.com/files/2292172",

                             keep_default_na=False, na_values=[""])

    species_df = pd.read_csv("https://ndownloader.figshare.com/files/3299483",

                             keep_default_na=False, na_values=[""])

    species_sub = pd.read_csv("https://raw.githubusercontent.com/TheJacksonLaboratory/python-ecology-lesson/gh-pages/data/speciesSubset.csv",

                             keep_default_na=False, na_values=[""])
```

```python
# Subset our full data into 2 smaller DataFrames
surveys_sub = surveys_df.head(10)
surveys_sub_last_10 = surveys_df[-10:]
surveys_sub_last_10
```

```python
surveys_sub
```

```python
surveys_sub_last_10 = surveys_sub_last_10.reset_index(drop=True)
surveys_sub_last_10
```

```python
# We can do simple combinations of our DataFrames using pd.concat()

vertical_stack = pd.concat([surveys_sub, surveys_sub_last_10])
vertical_stack
```

```python
# What is the axis anyway?
surveys_df[1, 2]
```

```python
horizontal_stack = pd.concat([surveys_sub, surveys_sub_last_10], axis=1)
horizontal_stack
```

```python
type(horizontal_stack)
```

```python
horizontal_stack.iloc[:,9]
```

```python
surveys_sub.index = [5, 6, 7, 8, 9, 10, 11, 12, 13, 15]
surveys_sub
```

## Joining DataFrames!

```python
# Let's see what our original data looks like

surveys_df.head(3)
```

```python
species_df.head(3)
```

```python
# Use the inner join to combine the two tables
merged_inner = pd.merge(left=surveys_sub,
                       right=species_sub,
                       left_on="species_id",
                       right_on="species_id")
```

```python
merged_inner
```

```python
surveys_sub["species_id"].unique()
```

```python
species_sub["species_id"].unique()
```

```python
merged_left = pd.merge(left=surveys_sub,
                      right=species_sub,
                      left_on="species_id",
                      right_on="species_id",
                      how="left")
```

```python
merged_left
```
